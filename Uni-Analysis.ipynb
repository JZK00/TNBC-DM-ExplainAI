```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import logit

# 读取数据集
train = pd.read_csv("E:/Experiments/YulanPeng/Wenwen/2023Run/DM/Experiments/Clinical/Figs-New/train.csv")

# 目标变量和特征变量列表
target = 'DM'  # 假设目标变量是'DM'
features = ['T_staging', 'N_staging', 'Clinical_staging', 'GPH', 'NC', 'RATH', 'CLN', 'PF', 'NPF', 'Enhancement', 'macrocalcification', 'LNM']

# 存储每个特征的p值
p_values = {}

# 对每个特征进行单变量逻辑回归分析
for feature in features:
    # 构建单变量模型
    formula = f"{target} ~ {feature}"
    model = logit(formula, data=train).fit()
    
    # 获取特征的p值
    p_values[feature] = model.pvalues[feature]

# 将p值按升序排序
p_values_sorted = sorted(p_values.items(), key=lambda x: x[1])

# 输出每个特征的p值，按p值排序
print("单变量分析结果 (按p值排序)：")
for feature, p_value in p_values_sorted:
    print(f"{feature}: p-value = {p_value}")

# 筛选显著特征（假设我们选择p < 0.05的特征）
significant_features = [feature for feature, p_value in p_values_sorted if p_value < 0.05]

print("\n筛选出的显著特征：")
print(significant_features)
```
